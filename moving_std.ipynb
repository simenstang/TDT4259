```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from datetime import timedelta
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('consumption_temp.csv')

df['time'] = pd.to_datetime(df['time'])

df.set_index('time', inplace=True)
```

```python
rolling_std = df.groupby('location')['consumption'].rolling(window='10D').std().unstack(level=0)

plt.figure(figsize=(18, 7))
for column in rolling_std.columns:
    rolling_std[column].plot(label=column)

plt.title('10-Day Moving Standard Deviation for Consumption by City')
plt.xlabel('Time')
plt.ylabel('Consumption Std Deviation')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
rolling_std = df.groupby('location')['temperature'].rolling(window='10D').std().unstack(level=0)

plt.figure(figsize=(18, 7))
for column in rolling_std.columns:
    rolling_std[column].plot(label=column)

plt.title('10-Day Moving Standard Deviation for temperature by City')
plt.xlabel('Time')
plt.ylabel('temperature Std Deviation')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
